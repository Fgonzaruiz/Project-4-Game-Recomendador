```python
df1 = pd.read_csv('rawg_columnas.csv')
df2= pd.read_csv('dfjuegos.csv')
```

```python
df1 = df1.reset_index()
df1
```

```python
df2 = df2.reset_index()
df2
```

```python
df1= df1[["level_0", "tags"]]
df1
```

```python
df_merged = df1.merge(df2, on=df1["level_0"])
```

```python
df_merged
```

```python
df_merged["tags"][0]
```

```python
df_merged['tag_names'] = df_merged['tags'].apply(lambda x: [tag['name'] for tag in eval(x)])
```

```python
df_merged["tag_names"][61030]
```

```python
import re
df_merged['tag_names'] = df_merged['tag_names'].apply(lambda x: [re.sub(r'[^a-zA-Z0-9\s]', '', item).strip() for item in x if item.strip() != ''])
```

```python
df_merged["tag_names"][61030]
```

```python
df_merged = df_merged.drop(['key_0', 'level_0_x', 'level_0_y', 'tags'], axis=1)
df_merged
```

```python
df_merged.info()
```

```python
df_merged
```

```python
from collections import Counter
all_tags = " ".join(df_merged["tag_names"].values)
words = all_tags.split()
word_counts = Counter(words)
word_counts

```

```python
import matplotlib.pyplot as plt
frequencies = list(word_counts.values())

plt.hist(frequencies, bins=50, log=True)
plt.xlabel('Frecuencia')
plt.ylabel('Cantidad de palabras')
plt.title('Histograma de frecuencia de palabras')
plt.show()
```

```python
umbral = 100
relevant_words = [word for word, count in word_counts.items() if count >= umbral]
df_merged = df_merged[df_merged["tag_names"].apply(lambda x: any(word in x for word in relevant_words))]
```

```python
import ast

for index, row in df_merged.iterrows():
    tag_names_str = row["tag_names"]
    tag_names_list = ast.literal_eval(tag_names_str)
    tag_names_str = ", ".join(tag_names_list)
    tag_names_str = tag_names_str.strip('[]')
    df_merged.loc[index, "tag_names_str"] = tag_names_str
```
