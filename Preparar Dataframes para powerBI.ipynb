Preparamos el dataset ya limpio y transformado para introducirlo en PowerBI y asi no tener que cambiar los datos a "mano":

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('df_streamlit.csv')
```

```python
df_generos = df[["index","name","Action","Adventure","Arcade","Board Games","Card","Casual","Educational","Family","Fighting","Indie","Massively Multiplayer","Platformer","Puzzle","RPG","Racing","Shooter","Simulation","Sports","Strategy"]]
df_generos[["Action","Adventure","Arcade","Board Games","Card","Casual","Educational","Family","Fighting","Indie","Massively Multiplayer","Platformer","Puzzle","RPG","Racing","Shooter","Simulation","Sports","Strategy"]] = df_generos[["Action","Adventure","Arcade","Board Games","Card","Casual","Educational","Family","Fighting","Indie","Massively Multiplayer","Platformer","Puzzle","RPG","Racing","Shooter","Simulation","Sports","Strategy"]].replace({1: True, 0: False})

df_generos
```

```python
df_plataforma = df[['index', 'name','Android', 'Apple II', 'Game Boy', 'Game Boy Advance', 'Game Boy Color','Linux', 'Nintendo 3DS', 'Nintendo 64', 'Nintendo DS','Nintendo Switch', 'PC', 'PSP', 'PlayStation', 'PlayStation 2','PlayStation 3', 'PlayStation 4', 'PlayStation 5', 'Web', 'Wii','Wii U', 'Xbox', 'Xbox 360', 'Xbox One', 'Xbox Series S/X', 'iOS','macOS']]
df_plataforma[['Android', 'Apple II', 'Game Boy', 'Game Boy Advance', 'Game Boy Color','Linux', 'Nintendo 3DS', 'Nintendo 64', 'Nintendo DS','Nintendo Switch', 'PC', 'PSP', 'PlayStation', 'PlayStation 2','PlayStation 3', 'PlayStation 4', 'PlayStation 5', 'Web', 'Wii','Wii U', 'Xbox', 'Xbox 360', 'Xbox One', 'Xbox Series S/X', 'iOS','macOS']] = df_plataforma[['Android', 'Apple II', 'Game Boy', 'Game Boy Advance', 'Game Boy Color','Linux', 'Nintendo 3DS', 'Nintendo 64', 'Nintendo DS','Nintendo Switch', 'PC', 'PSP', 'PlayStation', 'PlayStation 2','PlayStation 3', 'PlayStation 4', 'PlayStation 5', 'Web', 'Wii','Wii U', 'Xbox', 'Xbox 360', 'Xbox One', 'Xbox Series S/X', 'iOS','macOS']].replace({1: True, 0: False})
df_plataforma
```

```python
df_stores = df[['index','name','App Store',
       'Epic Games', 'GOG', 'Google Play', 'Nintendo Store',
       'PlayStation Store', 'Steam', 'Xbox 360 Store', 'Xbox Store']]
df_stores[['App Store',
       'Epic Games', 'GOG', 'Google Play', 'Nintendo Store',
       'PlayStation Store', 'Steam', 'Xbox 360 Store', 'Xbox Store']] = df_stores[['App Store',
       'Epic Games', 'GOG', 'Google Play', 'Nintendo Store',
       'PlayStation Store', 'Steam', 'Xbox 360 Store', 'Xbox Store']].replace({1: True, 0: False})
df_stores
```

```python
df_modojuego = df[['index' , 'name', 'Singleplayer', 'Multiplayer']]
df_modojuego[['Singleplayer', 'Multiplayer']] = df_modojuego[['Singleplayer', 'Multiplayer']].replace({1: True, 0: False})
df_modojuego
```

```python
df_melted = df_generos.melt(id_vars=['index', 'name'], var_name='Genero', value_name='Juego')
counts_df = df_melted.groupby(['Genero', 'Juego']).size().unstack(fill_value=0)
resultadoGenero = counts_df.reset_index().rename(columns={True: 'True', False: 'False'})
print(resultadoGenero[['Genero', 'True', 'False']])
```

```python
df_melted = df_stores.melt(id_vars=['index', 'name'], var_name='Stores', value_name='Tiendas')
counts_df = df_melted.groupby(['Stores', 'Tiendas']).size().unstack(fill_value=0)
resultadoStore = counts_df.reset_index().rename(columns={True: 'True', False: 'False'})
print(resultadoStore[['Stores', 'True', 'False']])
```

```python
df_melted = df_plataforma.melt(id_vars=['index', 'name'], var_name='plataforma', value_name='Nombre')
counts_df = df_melted.groupby(['plataforma', 'Nombre']).size().unstack(fill_value=0)
resultadoplata = counts_df.reset_index().rename(columns={True: 'True', False: 'False'})
print(resultadoplata[['plataforma', 'True', 'False']])
```

```python
df_melted = df_modojuego.melt(id_vars=['index', 'name'], var_name='modojuego', value_name='indice')
counts_df = df_melted.groupby(['modojuego', 'indice']).size().unstack(fill_value=0)
resultadomodojuego = counts_df.reset_index().rename(columns={True: 'True', False: 'False'})
print(resultadomodojuego[['modojuego', 'True', 'False']])
```

```python
resultadomodojuego.to_csv('modojuego.csv', index=False, encoding='utf-8')
resultadoplata.to_csv("plataforma.csv", index=False, encoding='utf-8')
resultadoStore.to_csv("stores.csv", index=False, encoding='utf-8')
resultadoGenero.to_csv("generos.csv", index=False, encoding='utf-8')
```
